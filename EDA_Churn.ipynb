# Exploratory Data Analysis (EDA) Report: Telco Customer Churn

This report summarizes the findings from the exploratory data analysis of the Telco Customer Churn dataset. The dataset contains 7,043 customer records with 21 attributes encompassing demographic information, services used, billing methods, and churn status.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')

sns.set_theme(style='whitegrid')

# Load dataset
df = pd.read_csv(r'C:\Users\gurja\OneDrive\Desktop\Ml Start\WA_Fn-UseC_-Telco-Customer-Churn.csv')
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
df.head()
```

## 1. Dataset Overview
- **Rows:** 7,043
- **Columns:** 21 (1 numerical, 2 continuous, 18 categorical)
- **Missing Values:** `TotalCharges` contains 11 missing values (which appeared as blanks). All other columns are complete.
- **Overall Churn Rate:** 26.54% of customers in the dataset churned, while 73.46% were retained.

> [!NOTE]
> The dataset is imbalanced towards non-churners (approx. 3:1 ratio). Any predictive model built on this data should account for this class imbalance (e.g., using SMOTE, class weights, or appropriate metrics like F1-score/AUC rather than simple accuracy).

```python
# Missing Values
print("Missing Values:\n", df.isnull().sum()[df.isnull().sum() > 0])

# Overall Churn Rate
plt.figure(figsize=(6, 6))
plt.pie(df['Churn'].value_counts(), labels=df['Churn'].value_counts().index, autopct='%1.2f%%', colors=['#66b3ff', '#ff9999'], startangle=90)
plt.title('Overall Churn Rate')
plt.show()
```

## 2. Key Insights & Drivers of Churn

Based on data profiling, we identified several strong indicators of customer churn.

### A. Contract Types & Tenure
- **Contract Type is a Massive Driver:** Customers on a `Month-to-month` contract have an exceptionally high churn rate of **42.7%**. In stark contrast, those on a `One year` contract churn at 11.3%, and those on a `Two year` contract have a mere 2.8% churn rate.
- **Tenure:** Churn happens early. The median tenure for churned customers is just **10 months**, compared to **38 months** for retained customers.

> [!TIP]
> Locking customers into longer contracts or providing incentives during the critical first 10 months can heavily reduce churn.

```python
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Contract Type
sns.countplot(data=df, x='Contract', hue='Churn', ax=axes[0], palette='Set2')
axes[0].set_title('Churn by Contract Type')

# Tenure
sns.boxplot(data=df, x='Churn', y='tenure', ax=axes[1], palette='Set2')
axes[1].set_title('Tenure vs Churn')

plt.tight_layout()
plt.show()
```

### B. Service Variations
- **Internet Service:** `Fiber optic` users are highly likely to churn (**41.9%** churn rate), whereas `DSL` users are much safer (19.0% churn rate). Customers without internet service rarely churn (7.4%).
- **Add-on Services:** Customers who **do not have** supplementary services like `OnlineSecurity`, `OnlineBackup`, `DeviceProtection`, and `TechSupport` have churn rates hovering around 40-42%. Conversely, those who subscribe to these services have their churn rates drop to ~15-22%.

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

# Internet Service
sns.countplot(data=df, x='InternetService', hue='Churn', ax=axes[0], palette='pastel')
axes[0].set_title('Churn by Internet Service')

# Online Security
sns.countplot(data=df, x='OnlineSecurity', hue='Churn', ax=axes[1], palette='pastel')
axes[1].set_title('Churn by Online Security')

# Tech Support
sns.countplot(data=df, x='TechSupport', hue='Churn', ax=axes[2], palette='pastel')
axes[2].set_title('Churn by Tech Support')

plt.tight_layout()
plt.show()
```

### C. Financial Factors & Billing
- **Payment Method:** The `Electronic check` payment method stands out dramatically with a **45.3% churn rate**. Other payment methods (Mailed check, Bank transfer, Credit card) all sit between 15% and 19%.
- **Monthly Charges:** Customers who churn tend to pay more per month on average (**$74.44**) compared to retained customers (**$61.27**). The high cost of Fiber Optic service might be a driving factor here.
- **Paperless Billing:** Interestingly, customers utilizing `PaperlessBilling` have a higher churn rate (33.6% vs 16.3%).

```python
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Payment Method
sns.countplot(data=df, y='PaymentMethod', hue='Churn', ax=axes[0], palette='muted')
axes[0].set_title('Churn by Payment Method')

# Monthly Charges
sns.kdeplot(data=df, x='MonthlyCharges', hue='Churn', fill=True, ax=axes[1], palette='muted')
axes[1].set_title('Density of Monthly Charges by Churn')

plt.tight_layout()
plt.show()
```

### D. Demographic Influences
- **Gender:** Gender has virtually no impact on churn (Male: 26.1%, Female: 26.9%).
- **Dependents & Partners:** Customers *without* dependents (31.3% churn) or partners (33.0% churn) are roughly twice as likely to churn compared to those with dependents (15.5%) or partners (19.7%).

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# Gender
sns.countplot(data=df, x='gender', hue='Churn', ax=axes[0], palette='deep')
axes[0].set_title('Churn by Gender')

# Dependents
sns.countplot(data=df, x='Dependents', hue='Churn', ax=axes[1], palette='deep')
axes[1].set_title('Churn by Dependents')

# Partner
sns.countplot(data=df, x='Partner', hue='Churn', ax=axes[2], palette='deep')
axes[2].set_title('Churn by Partner')

plt.tight_layout()
plt.show()
```

## 3. Final Conclusions & Recommendations

The dataset provides clear, actionable profiles of who is leaving the service. 

> [!IMPORTANT]
> **The High-Risk Customer Profile:**
> A new customer (under 10 months tenure), single/without dependents, on a month-to-month contract, using Fiber Optic internet without tech support or online security, and paying via Electronic Check.

**Recommendations:**
1. **Investigate Fiber Optic Services:** The 42% churn rate for Fiber optic users combined with their higher monthly charges suggests dissatisfaction. The company should investigate connectivity quality, service reliability, or pricing models for this specific product.
2. **Promote Long-term Contracts:** Incentivize month-to-month users to switch to 1-year or 2-year contracts.
3. **Bundle Add-ons:** Offer discounted bundles that include Tech Support and Online Security, as these significantly increase customer stickiness.
4. **Audit Electronic Checks:** Investigate the Electronic check payment gateway. A 45% churn rate indicates there may be hidden fees, recurring failures, or friction causing customers to cancel.
