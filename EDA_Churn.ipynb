# Telco Customer Churn - Exploratory Data Analysis
This notebook contains univariate and bivariate analysis, and determines the best features for predicting churn.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder

import warnings
warnings.filterwarnings('ignore')

sns.set_theme(style='darkgrid')

# Load dataset
df = pd.read_csv(r'C:\Users\gurja\OneDrive\Desktop\Ml Start\WA_Fn-UseC_-Telco-Customer-Churn.csv')
# Clean data
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
df.dropna(inplace=True)
df.head()
```

## 1. Univariate Analysis
Let's look at the distribution of some key numerical and categorical variables.

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 5))
sns.histplot(df['tenure'], bins=30, kde=True, ax=axes[0], color='skyblue')
axes[0].set_title('Distribution of Tenure')
sns.histplot(df['MonthlyCharges'], bins=30, kde=True, ax=axes[1], color='salmon')
axes[1].set_title('Distribution of Monthly Charges')
sns.histplot(df['TotalCharges'], bins=30, kde=True, ax=axes[2], color='lightgreen')
axes[2].set_title('Distribution of Total Charges')
plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
sns.countplot(data=df, x='gender', ax=axes[0, 0], palette='Set2')
axes[0, 0].set_title('Gender Distribution')
sns.countplot(data=df, x='Contract', ax=axes[0, 1], palette='Set2')
axes[0, 1].set_title('Contract Type Distribution')
sns.countplot(data=df, x='InternetService', ax=axes[1, 0], palette='Set2')
axes[1, 0].set_title('Internet Service Distribution')
sns.countplot(data=df, x='Churn', ax=axes[1, 1], palette='Set1')
axes[1, 1].set_title('Target Variable: Churn Distribution')
plt.tight_layout()
plt.show()
```

## 2. Bivariate Analysis
Analyzing how features relate to Customer Churn.

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
sns.boxplot(data=df, x='Churn', y='tenure', ax=axes[0], palette='Set1')
axes[0].set_title('Tenure vs Churn')
sns.boxplot(data=df, x='Churn', y='MonthlyCharges', ax=axes[1], palette='Set1')
axes[1].set_title('Monthly Charges vs Churn')
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
sns.countplot(data=df, x='Contract', hue='Churn', ax=axes[0], palette='Set1')
axes[0].set_title('Churn Rate by Contract Type')
sns.countplot(data=df, x='InternetService', hue='Churn', ax=axes[1], palette='Set1')
axes[1].set_title('Churn Rate by Internet Service')
plt.show()
```

## 3. Feature Importance
Using a Random Forest model to determine the most important predictors of churn.

```python
# Encode categorical variables
df_encoded = df.drop(columns=['customerID']).copy()
for col in df_encoded.select_dtypes(include=['object']).columns:
    le = LabelEncoder()
    df_encoded[col] = le.fit_transform(df_encoded[col])

X = df_encoded.drop('Churn', axis=1)
y = df_encoded['Churn']

# Train Random Forest
rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X, y)

# Plot Feature Importance
importances = rf.feature_importances_
indices = np.argsort(importances)[::-1]

plt.figure(figsize=(12, 6))
sns.barplot(x=importances[indices], y=X.columns[indices], palette='viridis')
plt.title('Feature Importances (Random Forest)')
plt.xlabel('Relative Importance')
plt.show()
```
